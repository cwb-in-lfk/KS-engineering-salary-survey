```python
import numpy as np
import pandas as pd

# open the file
sal_df = pd.read_csv('EmployeeData_2020.csv')



# ### This section looks for engineering jobs in general ###
# # Look for some kind engineer in the job title but not technician
e_df = sal_df[sal_df['Job_Title'].str.match('(.*gineer.*)|(.*Eng.*)|(.*eng.*)')==True]
e_df = e_df[e_df['Job_Title'].str.match('.*ician.*')==False]

# # Omit the universities, OITS (no "systems engineers"), legislature
e_df = e_df[e_df['Agency_Name'].str.match('(.*iversity.*)|(.*of Info.*)|(.*islature.*)|(.*Univ.*)')==False]
# ### END ###

# Look for just engineer associate (not PE)
e_df = e_df[e_df['Job_Title'].str.match('(.*sociate.*)|(.*EA.*)')==True]
e_df['EA'] = 'Engineering Associate'

# Omit the overtime pay - just report the base pay
e_df['regPay'] = e_df.Total_Gross_Pay - e_df.Overtime_Pay

#Omit entries where the annual pay is less than $40,000 (probably only a partial year)
greaterThan40K = e_df['regPay'] > 45000 # this is the boolean filter
e_df = e_df[greaterThan40K] # this is the filtered dataframe

print (e_df[['Agency_Name','Job_Title','regPay']].groupby(['Agency_Name','Job_Title']).aggregate(['mean','count']).astype('int').to_string(max_rows=500))
print ('\n\nSummary\n')
print (e_df.regPay.describe().astype('int'))
```

```python
def engPaySummary(df,i):   
    # ### This section looks for engineering jobs in general ###
    # # Look for some kind engineer in the job title but not technician
    e_df = df[df['Job_Title'].str.match('(.*gineer.*)|(.*Eng.*)|(.*eng.*)')==True]
    e_df = e_df[e_df['Job_Title'].str.match('.*ician.*')==False]

    # # Omit the universities, OITS (no "systems engineers"), legislature
    e_df = e_df[e_df['Agency_Name'].str.match('(.*iversity.*)|(.*of Info.*)|(.*islature.*)|(.*Univ.*)')==False]
    # ### END ###

    # Look for just engineer associate (not PE)
    e_df = e_df[e_df['Job_Title'].str.match('(.*sociate.*)|(.*EA.*)')==True]
    e_df['EA'] = 'Engineering Associate'

    # For the years that have it (2015-2020) Omit the overtime pay - just report the base pay
    if i in range(2015,2021): e_df['regPay'] = e_df.Total_Gross_Pay - e_df.Overtime_Pay

    #Omit entries where the annual pay is less than $40,000 (probably only a partial year)
    greaterThan40K = e_df['regPay'] > 45000 # this is the boolean filter
    e_df = e_df[greaterThan40K] # this is the filtered dataframe
    
    # remove extra whitespace so it looks pretty(ier)
    e_df['Agency_Name'] = e_df['Agency_Name'].str.strip()
    e_df['Job_Title'] = e_df['Job_Title'].str.strip()
#     e_df['Agency_Name'].str.strip()

    print (e_df[['Agency_Name','Job_Title','regPay']].groupby(['Agency_Name','Job_Title']).aggregate(['mean','count']).astype('int').to_string(max_rows=500))
    print ('\n\nSummary\n')
    print (e_df.regPay.describe().astype('int'))
    print ('__________________________________________________________________\n\n')
    
    
for i in range(2012,2021):
    fileName = 'EmployeeData_' + str(i) + '.csv'
    df = pd.read_csv(fileName)
    print (i)
#     print (df.columns)
#     print (df.describe())
#     print (df.head(n=25))
#     df['regPay'] = ['']
    if('Frequency' in df.columns):
        df.loc[df['Frequency'].str.match('.*Hourly.*')==True, 'regPay'] = df.Pay_Rate * 2080
        df.loc[df['Frequency'].str.match('.*Biweekly.*')==True, 'regPay'] = df.Pay_Rate * 26
#         print (df.head(n=25))
    engPaySummary(df, i)

        
```
